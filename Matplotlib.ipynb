```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,10)
```

```python
y = x**2
```

```python
plt.plot(x,y,'r-')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("weight")
plt.ylabel("Age")
plt.title("My Graph")
plt.legend(['Age'],loc=1)
```

# Multiple plots in a single Canvas

```python
plt.subplot(1,2,1) #rows,col,plotnumber
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
plt.subplot(221)
plt.plot(x,y,'r')

plt.subplot(222)
plt.plot(x,y,'b')

plt.subplot(223)
plt.plot(x,y,'y')

plt.subplot(224)
plt.plot(x,y,'g')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
for row in axes:
    for col in row:
        col.plot(x,y)
plt.tight_layout()
```
